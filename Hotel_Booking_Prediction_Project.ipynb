### Hypothesis Formulation questions

Hypothesis overall question: Which seasons of the year have more hotel stays? 

Support research: Summer is a great time to travel — except for the crowds and high temperatures. Sunny weather, long days, and exuberant nightlife turn Europe into a powerful magnet. This is followed by the closure of schools and universities.


**First hypothesis**

H0:Summer months are the peak season for hotel stays: μ = 3.580

Ha: Summer months are not the peak season for hotel stays μ =! 3.580


**Second Hypothesis**

H0: Hotel stays mean in the summer month is less than or equal to stays in the winter. μ <= 3.938

Ha:Hotel stays mean in the summer months is greater than stays in the winter: μ > 3.938







**Third hypothesis**

H0: The mean stays for summer months is less than or equal to the mean stays for the autumn months

Ha:The mean stays for summer months is greater than the mean stays for autumn months 

H0:mean1 </= mean3 Ha: mean1 > mean3 

**Fourth Hypothesis**

H0:The mean stays in the summer months is less than or equal to mean stays in the spring months

Ha: The mean stays in the summer months is greater than mean stays in the spring months

H0:mean1 </= mean4

Ha: mean1 > mean4

```
# visualization of country with % total bookings

country_notcanceled = hotel_notcanceled ['country']

def get_count(country_notcanceled, limit = None):
  if limit != None: 
    country_notcanceled = country_notcanceled.value_counts()[:limit]
  else:
    country_notcanceled = country_notcanceled.value_counts()

  x = country_notcanceled.index
  y = country_notcanceled/country_notcanceled.sum()*100

  return x.values,y.values
```

```
## For each country code select the country name

def plot(x, y, x_label=None,y_label=None, title=None, figsize=(7,5), type='bar'):

    sns.set_style('darkgrid')
    
    fig, ax = plt.subplots(figsize=figsize)
    
    if x_label != None:
        ax.set_xlabel(x_label)
    
    if y_label != None:
        ax.set_ylabel(y_label)
        
    if title != None:
        ax.set_title(title)
    
    if type == 'bar':
        sns.barplot(x,y, ax = ax)
    elif type == 'line':
        sns.lineplot(x,y, ax = ax)
        
    
    plt.show()
```

### **First hypothesis**

Null hypothesis: Summer months are the peak season for hotel stays: μ = 3.580

Alternative:Summer months are not the peak season for hotel stays μ =! 3.580

```
import scipy
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

population_mean = df['total_stays'].mean()
#z-score for summer
summer_mean = summer_df['total_stays'].mean()
n_summer = summer_df.shape[0]
summer_std_error = summer_df['total_stays'].std() / np.sqrt(n_summer)

#Find the z scores
summer_zscore = (summer_mean - population_mean) / summer_std_error
print('The z statistic for summer season is', summer_zscore)
print ('\n')

#Converting the obtained z-scores to p values
summer_pvalue = scipy.stats.norm.sf(abs(summer_zscore))*2
print('The pvalue of the summer season is', summer_pvalue)
print('\n')

#Test for significance at 5%
if summer_pvalue<0.05:
    print("indicates strong evidence against the null hypothesis, so you reject the null hypothesis")
else:
    print("indicates weak evidence against the null hypothesis, so you fail to reject the null hypothesis")
#Test for sensitivity at 1%
if summer_pvalue<0.01:
   print('Reject the null hypothesis, accept the alternative')
else:
  print('fail to reject the null hypothesis')
```
